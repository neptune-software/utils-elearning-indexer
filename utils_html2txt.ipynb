```python
import os
import re
from bs4 import BeautifulSoup

def check_and_create_directory(directory_name):
    if not os.path.exists(directory_name):
        os.makedirs(directory_name)
        print(f"Directory '{directory_name}' created successfully.")
    else:
        print(f"Directory '{directory_name}' already exists.")
        
def convert_html_to_text(html):
    # Parse HTML
    soup = BeautifulSoup(html, 'html.parser')
    # Extract text content
    text = soup.get_text()
    return text

def process_files(folder_path):
    # Traverse folder
    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)
        # Check if it's a txt file and filename starts with 8 digits
        if filename.endswith('.txt'):
            print(filename)
            with open(filepath, 'r', encoding='utf-8') as file:
                html_content = file.read()
            # Convert HTML content to plain text
            text_content = convert_html_to_text(html_content)
            # Build new filename
            new_filename = f"./data_ingested/text_{filename}"
            new_filepath = os.path.join(folder_path, new_filename)
            # Save as txt file
            with open(new_filename, 'w', encoding='utf-8') as file:
                file.write(text_content)
            print(f"Converted {filename} to {new_filename}")

# Specify folder path
folder_path = "/Users/gengsun/Documents/poc/eLearning/web-scraping-pptr/data"
check_and_create_directory("data_ingested")
# Process files
process_files(folder_path)
```
